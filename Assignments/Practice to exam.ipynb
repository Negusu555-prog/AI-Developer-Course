```python
# ⦁	Create a dataframe that shows how many different courses have been started in each month during 2023 year.
# convert_time_to_str 
# Assuming courses_copy is your DataFrame
courses_copy["course_start_date"] = pd.to_datetime(courses_copy["course_start_date"])
courses_copy["start_month"] = courses_copy["course_start_date"].dt.month

# Print the DataFrame to verify the changes
courses_copy
```

```python
# this is how many different courses have been started in each month during 2023 year
courses_in_each_month = courses_copy.groupby("start_month").size()
courses_in_each_month = courses_in_each_month.reset_index(name = "the_number_of_the_courses")
courses_in_each_month
```

```python
plt.plot(courses_in_each_month["start_month"],courses_in_each_month["the_number_of_the_courses"],marker='o', linestyle='--')
plt.title('chart that represent the trend of the number of opened courses during each month ')
plt.xlabel('Months')
plt.ylabel('Number of courses in each month')
plt.grid(True)
plt.show()
```
