```python
d = pd.read_csv(r"C:\Users\Negus\OneDrive\שולחן העבודה\Unsupervised Learning csvs\wine_dataset.csv") 
d
```

```python
d = d.drop(columns = "Unnamed: 0") 
d
```

```python
X = pd.get_dummies(d,drop_first = True)

scaler =StandardScaler()
scales_features = scaler.fit_transform(X)

ssd = []
for k in range(2,8):
    Kmeans = KMeans(n_clusters = k, random_state= 40, n_init=10)

    X["cluster"] = Kmeans.fit_predict(scales_features)

    cluster_centers = Kmeans.cluster_centers_ 
    
    ssd.append(Kmeans.inertia_)
ssd

plt.plot(range(2,8), ssd, 'o--')
plt.show()
```
